**SEABORN**

Seaborn is a powerful Python library for creating attractive, easy-to-read statistical data visualizations, built on top of Matplotlib. It simplifies plotting by offering high-level functions, built-in themes, and automatic dataset handling, making it ideal for exploring trends, distributions, and relationships in data

**1. Line Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var1=[1,2,3,4,5,6,7]
var2=[2,3,4,1,5,6,7]
 
sns.lineplot(x=var1,y=var2)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var1=[1,2,3,4,5,6,7]
var2=[2,3,4,1,5,6,7]

x=pd.DataFrame({"var1":var1,"var2":var2})

sns.lineplot(x="var1",y="var2",data=x)
plt.show()
```

```python
plt.plot(var1,var2)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="autumn",markers=["o",">"])
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="autumn",markers=["o",">"],dashes=True,legend="full")   #legend=auto, brief, full or boolean
plt.grid()
plt.title("Python")
plt.show()
```
